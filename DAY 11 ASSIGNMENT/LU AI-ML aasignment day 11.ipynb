# 1. correlation between attrition and distance from home

```python
from scipy.stats import pearsonr
stats,p=pearsonr(dataset.Attrition,dataset.DistanceFromHome)
print("correllation between Attrition and distanceFromHome is",stats,"and p value is",p)
if(p<0.05):
    print("reject null hypotesis")
else:
    print("accept null hypothesis")
```
